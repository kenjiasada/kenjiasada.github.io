```python
df_c = pd.read_csv('crimes_Tokyo.csv', encoding="SHIFT-JIS")
```

```python
df_c['city_name'] = df_c['市区町丁'].str.extract(r'^([^区市町村]+[区市町村])')
```

```python
df_c
```

```python
df_c2 = df_c.groupby('city_name').sum()
```

```python
df_c2
```

```python
df_c3 = df_c2.rename(index={'東村':'東村山市', '武蔵村':'武蔵村山市', '羽村':'羽村市',
                            '西多摩郡奥多摩町':'奥多摩町', '西多摩郡日の出町':'日の出町', '西多摩郡檜原村':'檜原村', '西多摩郡瑞穂町':'瑞穂町',
                           '三宅島三宅村':'三宅村', '八丈島八丈町':'八丈町'})
```

```python
df_c3 = df_c3.reset_index()
```

```python
df_c4 = df_c3[['city_name', '凶悪犯計', '粗暴犯計', '侵入窃盗計','非侵入窃盗計','非侵入窃盗万引き', 'その他詐欺']]
```

```python
df_c4
```

## Build DF of population

```python
df_p = pd.read_excel('pop_Tokyo.xls', encoding="SHIFT-JIS", header=[3,5])
```

```python
df_p.head()
```

```python
df_p.columns = list(map('_'.join, df_p.columns.values))
```

```python
df_p
```

```python
df_p.columns
```

```python
df_p1 = df_p[['地   域_Unnamed: 1_level_1', '平成30年１月１日現在_Unnamed: 2_level_1']]
```

```python
df_p2 = df_p1.iloc[2:72, :]
```

```python
df_p3 = df_p2.dropna()
```

```python
df_p4 = df_p3.rename(columns={'地   域_Unnamed: 1_level_1':'city', '平成30年１月１日現在_Unnamed: 2_level_1':'pop'})
```

```python
df_p4.loc[df_p4['city'].str.contains(r'　'), 'city'] = df_p4.loc[df_p4['city'].str.contains(r'　'), 'city'].str.replace('　','')
```

```python
df_p4
```

## Build income dataframe

```python
df_i = pd.read_csv('income_Tokyo.csv', names=('rank', 'city', 'income'))
```

```python
df_i['income'] = df_i.income.str.extract(r'(.*)円')
```

```python
df_i['income'] = df_i.income.str.replace('万','').astype(int)
```

```python
df_i.dtypes
```

```python
df_i.head()
```

## Merge population and crime counts

```python
 merged = df_p4.merge(df_c4, left_on='city', right_on='city_name', how='outer')
# 合致しないラベルを残すのがアウター
```

```python
 merged.head()
```

```python
merged2 = merged.merge(df_i, left_on='city', right_on='city', how='outer')
```

```python
merged3 = merged2.dropna()
```

```python
merged3.head()
```

```python
merged3['highly_violent'] = merged3['凶悪犯計'] / merged['pop'] * 100
merged3['violent'] = merged3['粗暴犯計'] / merged['pop'] * 100
merged3['theft'] = merged3['侵入窃盗計'] / merged['pop'] * 100
merged3['shoplifting'] = merged3['非侵入窃盗万引き'] / merged['pop'] * 100
merged3['fraud'] = merged3['その他詐欺'] / merged['pop'] * 100
```

```python
merged4 = merged3[['city', 'pop', 'rank', 'income', 'highly_violent', 'violent',
       'theft', 'shoplifting', 'fraud']]
```

```python
merged4.head(3)
```

## build the list of elderly people pop

```python
elderly = pd.read_excel('2015.xls', header=8)
```

```python
elderly2 = elderly.drop(elderly.columns[[0,1,2,3,4,5,6]], axis=1)
```

```python
elderly2.columns
```

```python
elderly3 = elderly2[['Unnamed: 7', '総数', '人口密度', '15歳未満\n人口', '15～64歳\n人口', '65歳以上\n人口',
                     '一般世帯数', 'Unnamed: 51', 'Unnamed: 52']]
```

```python
elderly3
```

```python
elderly4 = elderly3.rename(columns={'総数':'total_pop', '人口密度':'density',
                                    '15歳未満\n人口':'pop_u15', '15～64歳\n人口':'pop_15_64', '65歳以上\n人口':'pop_o65','Unnamed: 7': 'city', '一般世帯数':'total_family', 'Unnamed: 51':'old_single', 'Unnamed: 52': 'old_couple'})


```

```python
elderly5 = elderly4.drop(index=[0, 1])
```

```python
elderly5[elderly5.old_single == '-']
```

```python
elderly5[elderly5.old_couple == '-']
```

```python
elderly6 = elderly5.replace('-', 0)
```

```python
elderly6.dtypes
```

```python
elderly6['pct_pop_u15'] = elderly6.pop_u15 / elderly6.total_pop * 100
elderly6['pct_pop_15_64'] = elderly6.pop_15_64 / elderly6.total_pop * 100
elderly6['pct_pop_o65'] = elderly6.pop_o65 / elderly6.total_pop * 100


elderly6['pct_old_single'] = elderly6.old_single / elderly6.total_family * 100
elderly6['pct_old_couple'] = elderly6.old_couple / elderly6.total_family * 100
```

```python
elderly6
```

## Merge shoplift list and old people list

```python
merged5 = merged4.merge(elderly6, left_on='city', right_on='city')
```

```python
merged5.head()
```

```python
merged5.to_csv('shoplifting.csv', index=False)
```

```python
# Get rid of small villages in islands.
merged6 = merged5[merged5.shoplifting != 0]
```

```python
# Get rid of cities with have high income(cities at the center of Tokyo)
merged7 = merged6[merged6['income'] < 6000000]
```

```python
merged7
```

```python
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
matplotlib.rcParams['pdf.fonttype']=42
%matplotlib inline
import numpy as np
```

```python
fig, ax = plt.subplots(figsize=(8, 8))
ax.grid(True)
ax.set_xlim(0, 40)
ax.set_ylim(0, 2)

sns.regplot('pct_pop_o65', 'shoplifting', data=merged7, marker='o', line_kws={'color':'red', 'linewidth':1},
            scatter_kws={'color':'grey'})

plt.title('Coefficiency between population of elderlies and shoplifting rate')

plt.tight_layout()
plt.savefig('old_people_shoplifting.pdf')
```

```python
import statsmodels.api as sm
from statsmodels.sandbox.regression.predstd import wls_prediction_std
```

```python
x = merged7[['pct_pop_o65']]
X = sm.add_constant(x)
y = merged7['shoplifting']

model = sm.OLS(y, X)
result = model.fit()
result.summary()
```

```python
x = merged7['pct_pop_o65']
X = sm.add_constant(x)
y = merged7['shoplifting']

model = sm.OLS(y, X)
result = model.fit()
result.summary()
```
